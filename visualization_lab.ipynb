# Animation

Ниже написан пример, как можно анимировать некоторые процессы. Анимация представляет собой последовательность кадров и в каждом кадре происходит обновление графика.

В указанном примере в каждом кадре генерится точка из нормального распределения и наносится на scatterplot.

## Params

* `fig` - системо координат заданного размера с заданным диапазоном разметки осей
* `update` - функция обновления графика
* `frames` - количество кадров
* `interval` - временной промежуток между кадрами в [мс]
* `repeat` - повтор.

```python
from IPython.display import display, clear_output
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

def update(frame_number):
    sns.scatterplot(x=np.random.randn(1), y=np.random.randn(1), color='green')


ani = FuncAnimation(fig, update, frames=100, interval = 50, repeat=False)
HTML(ani.to_html5_video())
```

В данном примере в каждом кадре отрисовывается равномерное распределение точек в количестве 500. В каждой следующей итерации пространство очищается от точек предыдущего кадра.

```python
fig, ax = plt.subplots(figsize=(5, 5))
def update(frame_number):
    # Очистка графика
    ax.cla()
    # отрисовка графика
    ax.set_xlim(-1, 11)
    ax.set_ylim(-1, 11)
    sns.scatterplot(x=np.random.randint(1, 1000, 500)/100, y=np.random.randint(1, 1000, 500)/100, color='green')


ani = FuncAnimation(fig, update, frames=100, interval = 50, repeat=False)
HTML(ani.to_html5_video())
```
